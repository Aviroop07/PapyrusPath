```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv("Books.csv")
ratings = pd.read_csv("Ratings.csv")
users = pd.read_csv("Users.csv")
```

```python
book_rating = ratings.merge(books,on='ISBN')
book_rating.isnull().sum()
```

```python
grouped_data = book_rating.groupby('Book-Title')
book_vs_ratingCount = grouped_data.count()['Book-Rating'].reset_index()
book_vs_ratingCount.rename(columns={'Book-Rating':'No. of ratings'},inplace=True)
#book_vs_ratingCount
```

```python
book_vs_avgRating = grouped_data.mean('Book-Rating').reset_index()
book_vs_avgRating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
#book_vs_avgRating
```

```python
popularity_df = book_vs_ratingCount.merge(book_vs_avgRating,on="Book-Title")[['Book-Title','No. of ratings','avg_rating']]
#popularity_df
```

```python
popularity_df = popularity_df[popularity_df['No. of ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popularity_df = popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Publisher','Image-URL-M','avg_rating','No. of ratings']]
popularity_df.head()
```

```python
book_vs_ratingCount_min50 = book_vs_ratingCount[book_vs_ratingCount['No. of ratings']>=50]
#book_vs_ratingCount
```

```python
ratings['Book-Rating'].max()
ratings['Book-Rating'].min()
```

```python
x = book_rating.groupby('User-ID').count()['Book-Rating']>200
active_users = x[x].index
active_users
```

```python
filtered_rating = book_rating[book_rating['User-ID'].isin(active_users)]
# filtered_rating.head(1)
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data =[]
    for i in similar_items:
        item=[]
        temp_df = books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
        
```

```python
import pickle
pickle.dump(popularity_df,open('popular.pkl','wb'))
pickle.dump(pt,open('pt.pkl','wb'))
pickle.dump(books,open('books.pkl','wb'))
pickle.dump(similarity_score,open('similar.pkl','wb'))
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
def getAdequateValue(df,key):
    ratings = df[key].values.tolist()
    ratings = sorted(ratings)
    threshold = round(len(ratings)*(0.9))
    return ratings[threshold]
    
```

```python
threshold_rating = getAdequateValue(data,'num_ratings')
```

```python
refined_data = data[data['num_ratings']>=threshold_rating]
threshold_value = getAdequateValue(refined_data,'rating')
refined_data = data[data['rating']>=threshold_value]
topBooks = refined_data.sort_values('rating',ascending=False)
topBooks
```

```python
genres = data['genre'].drop_duplicates().values.tolist()
no_of_genre = len(genres)
```

```python
genre_wise_suggestions = []
for genre in genres:
    genre_wise_suggestions.append(data[data['genre']==genre])
    
for i in range(no_of_genre):
    threshold_rating = getAdequateValue(genre_wise_suggestions[i],'rating')
    genre_wise_suggestions[i] = genre_wise_suggestions[i][genre_wise_suggestions[i]['rating']>=threshold_rating]
    genre_wise_suggestions[i] = genre_wise_suggestions[i].sort_values('num_ratings',ascending=False).head(20)
    
```

```python
genre_suggestions = {}

for i in range(no_of_genre):
    genre_suggestions[genres[i]] = genre_wise_suggestions[i]
```
